## 3. Naver API

```python
import os
import sys
import urllib.request
```

```python
# 정보 입력
client_id = os.getenv("NAVER_API_SECRET_ID")
client_secret = os.getenv("NAVER_API_SECRET_KEY")

# 조건 입력
target_text = "총기"

# 검색어 인코딩
encText = urllib.parse.quote(target_text)

# url 최종 완성
url = "https://openapi.naver.com/v1/search/news.json?query=" + encText # JSON 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

# 요청 발신
response = urllib.request.urlopen(request)

# 응답 수신 및 확인 & 예외처리
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
# 정보 입력
client_id = os.getenv("NAVER_API_SECRET_ID")
client_secret = os.getenv("NAVER_API_SECRET_KEY")

# 조건 입력
target_text = "총기"

params = {
    "query" : target_text,
    "display" : 20
}

# url & Parameter 최종 완성
base_url = "https://openapi.naver.com/v1/search/news.json"
query_text = encText
url = f"{base_url}?query={query_text}"

headers = {
    "X-Naver-Client-Id" : client_id,
    "X-Naver-Client-Secret" : client_secret
}

# 요청 발신
response = requests.get(url, headers = headers, params = params)

# 응답 수신 및 확인 & 예외처리
rescode = response.status_code
if rescode == 200 :
    print(response.json())
else : 
    print("error occured. error code : ", rescode)
```

## 4. Youtube API

```python
from googleapiclient.discovery import build
```

```python
load_dotenv()

api_key = os.getenv("GOOGLE_CLOUD_API_KEY")
video_id = "7qXFLRRMbKw"

youtube = build("youtube", "v3", developerKey = api_key)

comments = []
next_page_token = None

while True :
    response = youtube.commentThreads().list(
        part = "snippet",
        videoId = video_id,
        maxResults = 100,
        pageToken = next_page_token,
        textFormat = "plainText"
    ).execute()

    for item in response["items"] :
        comment = item["snippet"]["topLevelComment"]["snippet"]
        comments.append(
            {
                "author" : comment["authorDisplayName"],
                "text" : comment["textDisplay"],
                "likeCount" : comment["likeCount"],
                "publishedAt" : comment["publishedAt"]
            }
        )

    next_page_token = response.get("nextPageToken")
    if not next_page_token :
        break

yotube_comment_data = pd.DataFrame(comments)
yotube_comment_data.head(10)
```
