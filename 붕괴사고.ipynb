```python
# 1. 괄호 포함 소괄호 내의 모든 문자 삭제
# 2. 스페이스바 split
# 3. (숫자 포함 데이터, 그 이후), (일반 문자열) 분리
# 4. 일반 문자열은 법정동 코드
# 5. 숫자/-(번,지) 외 문자 삭제
# 6. - 기준 번/지 분리, 없으면 공백
import re

def final_addr_parser(addr):
    if not isinstance(addr, str):
        return pd.Series([None, None, None])
    # 1. 괄호 및 괄호 내 문자 삭제
    addr_clean = re.sub(r'\([^)]*\)', '', addr).strip()
    # 2. split
    parts = addr_clean.split()
    # 3. 숫자 포함 데이터 찾기
    num_idx = None
    for i, part in enumerate(parts):
        if re.search(r'\d', part):
            num_idx = i
            break
    if num_idx is None:
        # 숫자 포함 데이터가 없으면 전부 법정동
        return pd.Series([' '.join(parts), '', ''])
    # 4. 법정동
    beopjeongdong = ' '.join(parts[:num_idx])
    # 5. 숫자/-(번,지) 외 문자 삭제
    num_part = re.sub(r'[^0-9\-]', '', parts[num_idx])
    # 6. - 기준 분리
    if '-' in num_part:
        bun, ji = num_part.split('-', 1)
    else:
        bun, ji = num_part, ''
    return pd.Series([beopjeongdong, bun, ji])

# 기존 컬럼 덮어쓰기
df[['법정동', '번', '지']] = df['ADDR'].apply(final_addr_parser)
# df[['ADDR', '법정동', '번', '지']].to_csv("data/temp_addr_trans.csv", encoding='utf-8-sig', index=False)
```
